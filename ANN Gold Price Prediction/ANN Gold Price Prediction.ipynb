```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Data Science Projects/ANN Gold Price Prediction/gold_price_data.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
df.columns
```

```python
X = df[['SPX', 'USO', 'SLV', 'EUR/USD']].values.copy()
y = df['GLD'].values.copy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
```

```python
scaled_X_train = min_max_scaler.fit_transform(X_train)
scaled_X_test = min_max_scaler.fit_transform(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(10, activation='relu', input_dim=4))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer='Adam')
```

```python
history = model.fit(scaled_X_train, y_train, epochs= 50, validation_split=0.1)
```

```python
y_predict = model.predict(scaled_X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```
