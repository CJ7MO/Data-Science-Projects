```python
df_object = df[['no_of_adults', 'no_of_children', 'no_of_weekend_nights', 'no_of_week_nights', 'type_of_meal_plan', 
                'room_type_reserved', 'no_of_previous_cancellations', 'avg_price_per_room','booking_status']].copy()

df_object
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
```

```python
df_object['type_of_meal_plan'] = label_encoder.fit_transform(df_object['type_of_meal_plan'])
df_object['room_type_reserved'] = label_encoder.fit_transform(df_object['room_type_reserved'])
df_object['booking_status'] = label_encoder.fit_transform(df_object['booking_status'])

df_object
```

```python
X = df_object[['no_of_adults', 'no_of_children', 'no_of_weekend_nights',
       'no_of_week_nights', 'type_of_meal_plan', 'room_type_reserved',
       'no_of_previous_cancellations', 'avg_price_per_room']].copy()

y=df_object['booking_status'].copy()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score

def model_result(prediction):
  print('The accuracy score from the model is: {}'.format(accuracy_score(y_test,prediction)))
```

```python
from sklearn.linear_model import  LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(scaled_X_train, y_train)
```

```python
log_predict = log_model.predict(scaled_X_test)
model_result(log_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

knn = KNeighborsClassifier()
k_values = list(range(1,30))
pipeline = Pipeline(steps=[('knn',knn)])
param_grid = {'knn__n_neighbors':k_values}
cv_classifier = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
cv_classifier.fit(scaled_X_train,y_train)

print(cv_classifier.best_params_)
```

```python
predict_gridknn = cv_classifier.predict(scaled_X_test)

model_result(predict_gridknn)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier()
n_estimators = [16,32,64]
max_features = [2,3]
bootstrap = [True,False]
oob_score = [True,False]

param_grid_rfc = {'n_estimators': n_estimators, 'max_features':max_features, 'bootstrap':bootstrap, 'oob_score':oob_score}

grid_rfc = GridSearchCV(rfc_model, param_grid_rfc)
grid_rfc.fit(scaled_X_train, y_train.values.ravel())

print(grid_rfc.best_params_)
```

```python
predict_rfc = grid_rfc.predict(scaled_X_test)
model_result(predict_rfc)
```
