```python
def extract_time(df,column):
  df[column+'_Hour']= df[column].dt.hour
  df[column+'_Minute']= df[column].dt.minute
  df.drop(column, axis=1, inplace=True)
  return df
```

```python
def flight_time (x):
  if x > 4 and x <= 8 :
    return 'Early morning' 
  elif x > 8 and x <= 12:
    return 'Morning'
  elif x > 12 and x <= 16:
    return 'Afternoon'
  elif x > 16 and x <= 20:
    return 'Evening'
  elif x > 20 and x <= 24:
    return 'Night'
  else:
    return 'Late Night'
```

```python
df['Dep_Time_Hour'].apply(flight_time).value_counts().plot(kind='bar')
```

```python
def preprocessing_duration (x):
  if 'h' not in x:
    x = '0h '+ x
  elif 'm' not in x:
    x = x + ' 0m'
  return x
```

```python
df['Duration'] = df['Duration'].apply(preprocessing_duration)
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
ml_model_rfr = RandomForestRegressor()
```

```python
model = ml_model_rfr.fit(X_train, y_train)
```

```python
y_pred = ml_model_rfr.predict(X_test)
y_pred
```

```python
def mape(y_true, y_pred):
  y_true, y_pred = np.array(y_true), np.array(y_pred)
  return np.mean(np.abs((y_true - y_pred) / y_true)*100)
```

```python
mape(y_test,y_pred)
```

```python
def predict(ml_model):
  model = ml_model_rfr.fit(X_train, y_train)
  print('Training score is: {}'.format(model.score(X_train, y_train)))
  predict = ml_model_rfr.predict(X_test)
  print('The predictions are: {}'.format(predict))
  print('-'*80)
  from sklearn import metrics
  r2_score = metrics.r2_score(y_test,predict)
  print('r2 score is: {}'.format(r2_score))
  print('-'*80)
  print('MSE : ', metrics.mean_squared_error(y_test,predict))
  print('-'*80)
  print('MAE : ', metrics.mean_absolute_error(y_test,predict))
  print('-'*80)
  print('MAPE : ', metrics.mean_absolute_percentage_error(y_test,predict))
  print('-'*80)
  print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test,predict)))
  print('-'*80)
  print('MAPE_: ', mape(y_test,predict))
  print('-'*80)
  sns.histplot(y_test-predict, kde=True)
```

```python
predict(RandomForestRegressor())
```
