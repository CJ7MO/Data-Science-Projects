```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
```

```python
df = datasets.load_iris()
X = df.data
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30)
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([('pca', PCA(n_components=2)), ('std', StandardScaler()), ('dt', DecisionTreeClassifier())], verbose=True)
```

```python
pipeline.fit(X_train, y_train)
```

```python
predict_pipeline = pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
def model_result(predictions):
  print(f'The accuracy score of the model is {accuracy_score(y_test,predictions)}')
```

```python
model_result(predict_pipeline)
```
